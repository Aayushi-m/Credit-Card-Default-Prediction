# DATA CLEANING


There are various anomalies in the dataset and we have tried to resolve them by combining or clubbing the undocumented feature values.We have done this in order to obtain a cleaner datset so as to improve our final training accuracy. We have refered to various research papers on our data set before doing it.

```python

from IPython.display import IFrame
IFrame('https://towardsdatascience.com/catching-a-welcher-classifying-a-credit-card-defaulter-f4b21547a618', width=800, height=400)
```

```python
df.loc[df.MARRIAGE == 0, 'MARRIAGE'] = 3
df.MARRIAGE.value_counts()
```

0 in Marriage --> can be categorized as 'other' 


FEATURE NAME-EDUCATION
0 is Undocumented 
5 & 6 are both labeled unknown 
CASE 1 - documenting 5  & 6 both as Other(Category - 4)
CASE 2 - documenting 5  & 6 both as Unknown(Category - 5)

```python
fil = (df.EDUCATION == 5) | (df.EDUCATION == 6) | (df.EDUCATION == 0)
df.loc[fil, 'EDUCATION'] = 4
df.EDUCATION.value_counts()
```

```python
df1=pd.read_csv('UCI_Credit_Card.csv')
```

```python
fil2 =  (df1.EDUCATION == 6) | (df.EDUCATION == 0)
df1.loc[fil2, 'EDUCATION'] = 5
df1.EDUCATION.value_counts()
```

In MARRIAGE also 0 is unlabelled so we can put it under category 3 only which is others.

```python
df.loc[df1.MARRIAGE == 0, 'MARRIAGE'] = 3
df.MARRIAGE.value_counts()
```

```python
fil = (df.PAY_1 == -2) | (df.PAY_1 == -1) | (df.PAY_1 == 0)
df.loc[fil, 'PAY_1'] = 0
fil = (df.PAY_2 == -2) | (df.PAY_2 == -1) | (df.PAY_2 == 0)
df.loc[fil, 'PAY_2'] = 0
fil = (df.PAY_3 == -2) | (df.PAY_3 == -1) | (df.PAY_3 == 0)
df.loc[fil, 'PAY_3'] = 0
fil = (df.PAY_4 == -2) | (df.PAY_4 == -1) | (df.PAY_4 == 0)
df.loc[fil, 'PAY_4'] = 0
fil = (df.PAY_5 == -2) | (df.PAY_5 == -1) | (df.PAY_5 == 0)
df.loc[fil, 'PAY_5'] = 0
fil = (df.PAY_6 == -2) | (df.PAY_6 == -1) | (df.PAY_6 == 0)
df.loc[fil, 'PAY_6'] = 0
```

```python
def corr(Col1, Col2):
    res = df.groupby([Col1, Col2]).size().unstack()
    res['perc'] = (res[res.columns[1]]/(res[res.columns[0]] + res[res.columns[1]]))
    return res

#In this function we are particularly seeing the percentage of default among all.
```
